```python
#Plotting Example Of Linear Data
import matplotlib.pyplot as plt
X = [1, 2, 3, 4, 5]
Y = [5, 6, 7, 8, 9]
plt.figure(figsize=(10,7))
plt.scatter(X, Y, c="r")
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.savefig("Linear.png")
plt.show()
```

```python
#Making Circular Dataset Using sklearn's make_circles
from sklearn.datasets import make_circles

n_samples = 1000

X, Y = make_circles(n_samples, random_state=42, noise=0.03)
```

```python
#Checking the shapes of X(features) and Y(labels)
X.shape, Y.shape
```

```python
#Creating Dataframe using X and Y
import pandas as pd
import tensorflow as tf

circles = pd.DataFrame({"X0":X[:,0],
                        "X1":X[:,1],
                        "label":Y})

circles
```

```python
#Plotting and Saving Circular Dataset
plt.figure(figsize=(10,7))
plt.scatter(X[:,0], X[:,1], c=Y, cmap=plt.cm.RdYlBu)
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.title("Non Linear Data")
plt.savefig("NonLinear.png")
plt.show()
```

```python
#Creating Dataset to implement activation function on.
import numpy as np

X = np.arange(-20, 20, 1)
X
```

```python
#Calculating and Plotting for Linear Activation

def linear(X):
    return X

X = linear(X)
plt.figure(figsize=(10, 7))
plt.plot(X)
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.title("Linear Activation Function")
plt.savefig("LinearActivation.png")
plt.show()
```

```python
#Calculating and Plotting for Sigmoid Activation

def sigmoid(X):
    return (1/(1 + np.exp(-X)))

X = sigmoid(X)
plt.figure(figsize=(10, 7))
plt.plot(X)
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.title("Sigmoid Activation Function")
plt.savefig("SigmoudActivation.png")
plt.show()
```

```python
#Calculating and Plotting for ReLU activation

X = tf.range(-20, 20)
X = tf.keras.activations.relu(X).numpy()
plt.figure(figsize=(10, 7))
plt.plot(X)
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.title("ReLU Activation Function")
plt.savefig("ReLUActivation.png")
plt.show()
```
